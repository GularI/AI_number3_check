```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Load data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Build model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Evaluate model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")

# Plot training history
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training Progress')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Predict a sample
sample = np.expand_dims(x_test[0], axis=0)
prediction = model.predict(sample)
print("Predicted digit:", prediction.argmax())
plt.imshow(x_test[0], cmap='gray')
plt.title(f"Predicted: {prediction.argmax()}")
plt.show()
```

```python
# Step 0: Import tools we need
import tensorflow as tf            # tool to make and train the AI model
from tensorflow.keras import layers, models  # building blocks for the model
import matplotlib.pyplot as plt    # to draw graphs and show images
import numpy as np                 # for working with numbers and arrays
from PIL import Image              # to open your own image
from google.colab import files     # to upload files from your computer to Colab

# Step 1: Upload your image to Colab
uploaded = files.upload()  # click "Choose File" and pick Number3.png from your Desktop

# Step 2: Load MNIST dataset (digits 0-9) for training
# This dataset has thousands of small images of handwritten numbers
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Step 3: Make pixel values smaller (0 to 1 instead of 0 to 255)
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 4: Build the AI model (a small neural network)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # flatten image into a line of 784 numbers
    layers.Dense(128, activation='relu'),  # first layer with 128 “neurons”
    layers.Dropout(0.2),                   # randomly ignore some neurons while training
    layers.Dense(10, activation='softmax') # last layer predicts 0-9
])

# Step 5: Tell the model how to learn
model.compile(
    optimizer='adam',                       # how the model changes itself
    loss='sparse_categorical_crossentropy', # how the model measures mistakes
    metrics=['accuracy']                     # what we check while training
)

# Step 6: Train the model
history = model.fit(
    x_train, y_train,          # the training images and their numbers
    epochs=5,                  # how many times the model looks at all images
    validation_data=(x_test, y_test)  # test after each epoch to see accuracy
)

# Step 7: Check model accuracy on test set
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc:.4f}")  # shows how good the model is

# Step 8: Draw a graph of training progress
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Step 9: Open your own image
image_path = "Number3.png"                  # uploaded file name
#img = Image.open(image_path).convert('L')   # make it black & white

# Step 10: Resize and normalize your image
img = img.resize((28, 28))          # resize to same size as MNIST images
img_array = np.array(img) / 255.0   # make pixel values between 0 and 1
img_array = 1 - img_array           # invert colors if needed (white digit on black)

# Step 11: Prepare image for model
sample = np.expand_dims(img_array, axis=0)  # add one dimension so model can read it

# Step 12: Let the model guess the number
prediction = model.predict(sample)
print("Predicted digit:", prediction.argmax())  # shows the number it thinks

# Step 13: Show the image
plt.imshow(img_array, cmap='gray')
plt.title(f"Predicted: {prediction.argmax()}")
plt.show()
```
